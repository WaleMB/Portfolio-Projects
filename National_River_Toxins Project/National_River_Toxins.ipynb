```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset to a dataframe
df = pd.read_csv('National_River_Toxin_Dataset_1.csv')
df.head()
```

```python
#To get the information about the Dataframe
df.info()
```

```python
#Show number of unique values in each column
df.nunique()
```

```python
#Check the count of the Null values in each column
df.isnull().sum()
```

```python
#fill missing values in numeric columns
numeric_columns= df.select_dtypes(include= [np.number]).columns
df[numeric_columns] = df[numeric_columns].fillna(df[numeric_columns].mean())
df[numeric_columns]
```

```python
#Show unique values in single column
df['River_System'].unique()
```

```python
#Description of the data in the DataFrame
df.describe()
```

```python
#Convert Date to Datetime datatype
df['Date'] = pd.to_datetime(df['Date'])
df.info()
```

```python
#Show the barchart of the Mercury level of the River system
xy=sns.barplot(x= 'River_System', y = 'Mercury', data=df, palette='tab20c', hue='River_System' , err_kws={'linewidth': 0})
plt.ylabel('Average Mercury Level')
plt.xlabel('River System')
plt.title('Average mercury level of the River system')
for c in xy.containers:
    xy.bar_label(c,)
```

```python
#Drop date field from dataframe
df2= df.drop(columns = 'Date')
```

```python
#Shows maximum values River system data
df2.groupby('River_System').max()
```

```python
#Shows minimum values River system data
df2.groupby('River_System').min()
```

```python
#show mean values of River system toxin data
mean_tx = df2.groupby('River_System').mean()
df_mean = round(mean_tx.reset_index(), 2)
df_mean
```

```python
#Show Stacked Barchart of Toxin mean levels
df3 = df_mean.plot.bar(x='River_System', stacked=True, title='Toxin Levels', xlabel= 'River_System', ylabel = 'Mean')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
for c in df3.containers:

    # Optional: if the segment is small or 0, customize the labels
    labels = [v.get_height() if v.get_height() > 0 else '' for v in c]
    
    # remove the labels parameter if it's not needed for customized labels
    df3.bar_label(c, label_type='center')
```

```python
#Exploring Amazon data
Amazon = df[df['River_System'] == 'Amazon']
Amazon.head()
```

```python
#Arsenic levels over the years
sns.lineplot(data=Amazon, x='Date', y='Arsenic', label='Amazon', color='blue')
plt.show()
```

```python
#Exploring Toxin levels for 2018
fdf = df.loc[(df['Date'] >= '2018-01-01')
                     & (df['Date'] < '2018-12-31')]
fdf.head()
```

```python
#Create correlation Matrix
corr_matrix = round(df[numeric_columns].corr(), 2)
corr_matrix
```

```python
#Create heatmap
sns.heatmap(corr_matrix, annot=True, cmap="YlOrBr")
plt.show()
```

```python
Amazon_lead = df[df['River_System'] == 'Amazon']['Lead']
```

```python
Yangtze_lead = df[df['River_System'] == 'Yangtze']['Lead']
```

```python
from scipy import stats
#Calculate t-test and p-value
t_test, p_value = stats.ttest_ind(Amazon_lead, Yangtze_lead)
print(f"T-test between Amazon and Yangtze lead level: T-statistic {t_test}, P-value {p_value}")
```

```python
#Calculate linear regression analysis for Arsenic and Mercury toxins
from sklearn.linear_model import LinearRegression
X = df[['Arsenic']]
Y = df['Mercury']
model = LinearRegression()
model.fit(X, Y)
print(f"Linear Regression Coefficient: {model.coef_}")
print(f"Linear Regression Intercept: {model.intercept_}")
```

```python
sns.lmplot(x='Lead', y='Arsenic', data=df,
           scatter_kws={'alpha':0.3})
```

```python
# Calculate R-squared score
model.coef_
r2_score = model.score(X, Y)
print(f"Linear Regression Coefficient: {model.coef_}")
print(f"R-squared Score: {r2_score:.4f}")
```

```python
#Make predictions
y_pred = model.predict(X)

#Create a scatter plot with our regression line
plt.scatter(X, Y, color='b', alpha=0.5, label='Actual data')
plt.plot(X, y_pred, color='k', label='Regression Line')
plt.xlabel('Arsenic')
plt.ylabel('Mercury')
plt.title('Arsenic vs Mercury')
plt.legend()
plt.show()
```
